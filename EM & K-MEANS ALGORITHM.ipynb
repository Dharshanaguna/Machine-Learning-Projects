```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score,confusion_matrix
import pandas as pd
import numpy as np
```

```python
#load iris dataset
iris = datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data)
X.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']

y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
#fitting of model
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
plt.figure(figsize=(14,7))
```

```python
colormap = np.array(['red', 'lime', 'black'])
```

```python
# Plot the Original Classifications
plt.subplot(1, 2, 1)
```

```python
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
# Plot the Models Classifications
plt.subplot(1, 2, 2)
```

```python
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('K Mean Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
print('The accuracy score of K-Mean: ',accuracy_score(y, model.labels_))
```

```python
print('The Confusion matrixof K-Mean: ',confusion_matrix(y, model.labels_))
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(X)
xsa = scaler.transform(X)
xs = pd.DataFrame(xsa, columns = X.columns)
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=3)
gmm.fit(xs)

y_gmm = gmm.predict(xs)
```

```python
plt.subplot(2, 2, 3)
```

```python
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y_gmm], s=40)
plt.title('GMM Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
print('The accuracy score of EM: ',accuracy_score(y, y_gmm))
```

```python
print('The Confusion matrix of EM: ',confusion_matrix(y, y_gmm))
```
